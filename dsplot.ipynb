```python
from graphviz import Digraph, Graph

```

```python
dot = Graph(comment='The Round Table', node_attr={'shape': 'record'})
dot.node('A', 'King Arthur|X|Y')
dot.node('B', 'X|{Sir|Bedevere|Wise}|{{X|2}|Y|Z|W}', {'shape': 'record'})
dot.node('L', 'Sir Lancelot the Brave')
dot.edges(['AB', 'AL'])
dot.edge('B', 'L', constraint='false')
dot
```
